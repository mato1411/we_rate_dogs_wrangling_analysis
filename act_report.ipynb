# We Rate Dogs Report

This report describes gained insights about the top dog breeds rated by [WeRateDogs](https://twitter.com/dog_rates9). 
The report is based on the analysis of data a WeRateDogs Twitter archive and the information about the predicted breed of dog present in each tweets image according to a neural network. Additionally, the Twitter API was used to determine the number of retweet and favorites (likes) of each tweet.

WeRateDogs is using a very unique rating system which is a big part of it's popularity. That's why rating numerators are greater than the denominators and there is not fix range of allowed numerators and denominators. To compare the tweets using the rating, rating was normalized by dividing the numerators by the denominators.
